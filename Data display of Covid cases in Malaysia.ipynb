### Displaying Covid-19 cases data

```python
# importing all the libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# reading and transfering data in .csv file to Pandas dataframe (e.g. 'cases_state.csv')
state_cases = pd.read_csv('cases_state.csv')
```

```python
# Displaying data in dataframe
state_cases
```

```python
# Selecting data for certain state (e.g. Selangor)
cases_Selangor = state_cases[state_cases.values=='Selangor']
cases_Selangor
```

```python
# Selecting data for W.P. Kuala Lumpur
cases_WPKL = state_cases[state_cases.values=='W.P. Kuala Lumpur']
cases_WPKL
```

```python
# Setting date column as index and simplifying the dataframe names
dfSelangor = cases_Selangor.loc[:,['date','cases_new']].set_index('date')
dfWPKL = cases_WPKL.loc[:,['date','cases_new']].set_index('date')
```

```python
# Plotting cumulative cases for Selangor with 'date' set as index
%matplotlib inline

dfSelangor.cumsum().plot()
_=plt.xticks(rotation=45) 
```

```python
# Plotting daily cases for Selangor with 'date' set as index
%matplotlib inline

dfSelangor.plot()
_=plt.xticks(rotation=45) 
```

```python
# Plotting daily cases for Selangor with 'date' set as index
%matplotlib inline

ax = dfSelangor.plot()
dfWPKL.plot(ax=ax)

_=plt.xticks(rotation=45)  

ax.legend(['Selangor', "W.P. Kuala Lumpur"])
```
